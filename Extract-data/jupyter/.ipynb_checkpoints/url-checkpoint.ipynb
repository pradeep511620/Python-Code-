```python
import time
from datetime import datetime
from itertools import zip_longest
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
login_url ='https://www.trustpilot.com/review/raptorsupplies.com?page=1'
driver = webdriver.Chrome()
driver.maximize_window()
driver.get(login_url)
time.sleep(5)
```

```python
print(driver.page_source)
```

```python
Review_name_elements = [Re_name.text.strip() for Re_name in driver.find_elements(By.CSS_SELECTOR, 'span[data-consumer-name-typography]')]
Review_Title_elements = [h2.text.strip() for h2 in driver.find_elements(By.CSS_SELECTOR, "div.styles_reviewContent__0Q2Tg h2")]
Review_text_elements = [Re.text.strip().replace("\n", "") for Re in driver.find_elements(By.CSS_SELECTOR, "p[data-service-review-text-typography]")]
Review_stars_elements = [sr.get_attribute('src') for sr in driver.find_elements(By.CSS_SELECTOR, 'div[data-service-review-rating] img')]
Cus_count_elements = [count_re.text.strip().split(' ')[0] for count_re in driver.find_elements(By.CSS_SELECTOR, "span[data-consumer-reviews-count-typography]")]
Review_Country_elements = [count.text.strip() for count in driver.find_elements(By.CSS_SELECTOR, "div[data-consumer-country-typography] span")]
Last_time_elements = [dates.text.strip() for dates in driver.find_elements(By.CSS_SELECTOR, "section.styles_reviewContentwrapper__zH_9M time")]
print('Done')
values_list = []
for name, title,texts, starts, cus_count,country, times in zip_longest(Review_name_elements, Review_Title_elements, Review_text_elements, Review_stars_elements, Cus_count_elements, Review_Country_elements, Last_time_elements):
    # print(f"First Name: {name}, Title: {title}, Text: {texts}, Stars: {starts}, Country: {country}, Time: {times}")
    values_list.append({
        'Review_Name': name,
        'Review_Title': title,
        'Review_Text': texts,
        'Review_Stars': starts,
        'Review_Country': country,
        'Review_Time': times
    })

# Create a DataFrame
df = pd.DataFrame(values_list)

# Save to CSV
df.to_csv('reviews_data.csv', index=False)
    
```
